```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
movies=movies.head(10000)
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer as cvr
```

```python
cv=cvr(max_features=1000,stop_words='english')
```

```python
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
 from sklearn.metrics.pairwise import cosine_similarity
```

```python
similiarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
        mindex=movies[movies['title']==movie].index[0]
        distances=similiarity[mindex]
        mlist=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
        for i in mlist:
            print(movies.iloc[i[0]].title)
        
```

```python
recommend('Fast X')
```
